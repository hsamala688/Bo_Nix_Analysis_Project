```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("Bo_Nix_Stats_2024.csv")

plt.figure(figsize=(10,6))
sns.barplot(data=df, x='Opp', y='Cmp')
plt.xticks(rotation=45)
plt.title("Bo Nix Completion % per Game")
plt.ylabel("Completion %")
plt.xlabel("Opponent")
plt.show()
```
